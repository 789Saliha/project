```python
# ===============================
# Phase 1: Data Preparation & Cleaning
# ===============================

import pandas as pd
import numpy as np

# ---------------------------
# 1. Load raw dataset
# ---------------------------
df = pd.read_csv("C:\\Users\\zeeshan\\Downloads\\airlines_reviews.csv")

print("✅ Raw dataset loaded")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head(3))

# ---------------------------
# 2. Data Cleaning
# ---------------------------

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Convert Recommended to binary (Yes=1, No=0), handle missing values
df["Recommended"] = (
    df["Recommended"]
    .astype(str).str.strip().str.lower()
    .map({"yes": 1, "no": 0})
    .fillna(0).astype(int)
)

# Standardize Class and Type of Traveller
df["Class"] = df["Class"].astype(str).str.title().fillna("Unknown")
df["Type of Traveller"] = df["Type of Traveller"].astype(str).str.title().fillna("Unknown")

# Convert Review Date to datetime
df["Review Date"] = pd.to_datetime(df["Review Date"], errors="coerce")

# Drop rows with missing Review Date
df = df.dropna(subset=["Review Date"])

# ---------------------------
# 3. Feature Engineering
# ---------------------------

# Extract Year and Month from Review Date
df["Review_Year"] = df["Review Date"].dt.year
df["Review_Month"] = df["Review Date"].dt.month

# Map Month Flown to Season
def get_season(month):
    if pd.isna(month):
        return "Unknown"
    month = str(month).strip().title()
    if month in ["December", "January", "February"]:
        return "Winter"
    elif month in ["March", "April", "May"]:
        return "Spring"
    elif month in ["June", "July", "August"]:
        return "Summer"
    elif month in ["September", "October", "November"]:
        return "Autumn"
    return "Unknown"

df["Season"] = df["Month Flown"].apply(get_season)

# Determine Route Type: Domestic vs International
def get_route_type(route):
    if isinstance(route, str) and " to " in route:
        parts = [p.strip() for p in route.split(" to ")]
        if len(parts) == 2 and parts[0] == parts[1]:
            return "Domestic"
        return "International"
    return "Unknown"

df["Route_Type"] = df["Route"].apply(get_route_type)

# ---------------------------
# 4. Save cleaned dataset
# ---------------------------


print("\n✅ Cleaned + engineered dataset saved as airlines_reviews_clean.csv")
print("Sample rows:")
print(df.head(5))

print("\nColumn data types:")
print(df.dtypes)

print("\nMissing values per column:")
print(df.isna().sum())
```

```python
# ===============================
# Phase 2: Rule-based Recommendation System (with proxy Recommended)
# ===============================

import pandas as pd
import numpy as np

# ---------------------------
# 1. Create proxy "Recommended" column
# Rule: if Overall Rating >= 7 → Recommended (1), else 0
# ---------------------------
df["Recommended"] = (df["Overall Rating"] >= 7).astype(int)

# ---------------------------
# 2. Function: Recommend Airlines
# ---------------------------
def recommend_airlines(route=None, travel_class=None, traveller_type=None, top_n=5):
    """
    Smarter recommender:
    Filters dataset by route/class/traveller type (optional),
    then ranks airlines by a weighted score:
    (Avg_Rating * Recommend_Rate) * log(Review_Count+1)
    Returns top N airlines.
    """

    subset = df.copy()

    # Apply filters if provided
    if route:
        subset = subset[subset["Route"].str.contains(route, case=False, na=False)]
    if travel_class:
        subset = subset[subset["Class"].str.contains(travel_class, case=False, na=False)]
    if traveller_type:
        subset = subset[subset["Type of Traveller"].str.contains(traveller_type, case=False, na=False)]

    if subset.empty:
        return "⚠️ No matching reviews found. Try broader filters."

    # Group by Airline and compute stats
    summary = subset.groupby("Airline").agg(
        Avg_Rating=("Overall Rating", "mean"),
        Recommend_Rate=("Recommended", "mean"),
        Review_Count=("Airline", "count")
    ).reset_index()

    # Weighted score
    summary["Weighted_Score"] = (summary["Avg_Rating"] * summary["Recommend_Rate"]) * np.log1p(summary["Review_Count"])

    # Sort and return top N
    summary = summary.sort_values(by=["Weighted_Score", "Review_Count"], ascending=[False, False]).head(top_n)

    return summary

# ---------------------------
# 3. Example queries
# ---------------------------
print("Example 1: Economy Class flights Singapore → London")
print(recommend_airlines(route="Singapore to London", travel_class="Economy Class", top_n=3))

print("\nExample 2: Business Class (any route)")
print(recommend_airlines(travel_class="Business Class", top_n=3))

print("\nExample 3: Solo Leisure travellers to Singapore")
print(recommend_airlines(route="Singapore", traveller_type="Solo Leisure", top_n=3))
```

```python
# ===============================
# Phase 3: Personalized Scoring Recommendation System
# ===============================

import pandas as pd

# ---------------------------
# Function: Personalized Recommender
# ---------------------------
def personalized_recommend(route=None, travel_class=None, traveller_type=None, top_n=5,
                           w_rating=0.5, w_recommend=0.3, w_review_count=0.2):
    """
    Personalized scoring recommender:
    Combines Avg Rating, Recommend Rate, and Review Count
    into a weighted score.
    """

    subset = df.copy()

    # Apply flexible filters
    if route:
        subset = subset[subset["Route"].str.contains(route, case=False, na=False)]
    if travel_class:
        subset = subset[subset["Class"].str.contains(travel_class, case=False, na=False)]
    if traveller_type:
        subset = subset[subset["Type of Traveller"].str.contains(traveller_type, case=False, na=False)]

    if subset.empty:
        return "⚠️ No matching reviews found. Try broader filters."

    # Group by Airline
    summary = subset.groupby("Airline").agg(
        Avg_Rating=("Overall Rating", "mean"),
        Recommend_Rate=("Recommended", "mean"),
        Review_Count=("Airline", "count")
    ).reset_index()

    # Normalize review counts
    summary["Review_Count_Norm"] = summary["Review_Count"] / summary["Review_Count"].max()

    # Personalized score formula
    summary["Personalized_Score"] = (
        w_rating * summary["Avg_Rating"] +
        w_recommend * summary["Recommend_Rate"] * 10 +  # scale to match rating range
        w_review_count * summary["Review_Count_Norm"] * 10
    )

    # Sort by personalized score
    summary = summary.sort_values(by="Personalized_Score", ascending=False).head(top_n)

    return summary

# ---------------------------
# Example queries
# ---------------------------
print("\nExample 1: Economy Class flights Singapore → London")
print(personalized_recommend(route="Singapore to London", travel_class="Economy Class", top_n=3))

print("\nExample 2: Business Class (any route)")
print(personalized_recommend(travel_class="Business Class", top_n=3))

print("\nExample 3: Solo Leisure travellers to Singapore")
print(personalized_recommend(route="Singapore", traveller_type="Solo Leisure", top_n=3))
```

```python
# ===============================
# Phase 5: Hybrid Recommendations (Streamlit-friendly)
# ===============================
import pandas as pd
import numpy as np

# Load precomputed SVD predictions
svd_pred_df = pd.read_csv("airline_svd_predictions.csv")

def hybrid_recommend(traveller_type, top_n=5):
    """
    Hybrid recommender using precomputed SVD scores:
    Combines rule-based scoring and ML-predicted SVD ratings.
    """

    # --- Rule-based part ---
    summary = df.groupby("Airline").agg(
        Avg_Rating=("Overall Rating", "mean"),
        Recommend_Rate=("Recommended", "mean"),
        Review_Count=("Airline", "count")
    ).reset_index()

    # Normalize review count
    summary["Review_Count_Norm"] = summary["Review_Count"] / summary["Review_Count"].max()

    # Rule-based score
    summary["Rule_Score"] = (
        0.5 * summary["Avg_Rating"] +
        0.3 * summary["Recommend_Rate"] * 10 +
        0.2 * summary["Review_Count_Norm"] * 10
    )

    # --- Add ML scores from precomputed CSV ---
    ml_scores = svd_pred_df[svd_pred_df["Traveller_Type"] == traveller_type][["Airline", "SVD_Score"]]
    summary = summary.merge(ml_scores, on="Airline", how="left").fillna(summary["Avg_Rating"])

    # --- Combine scores ---
    summary["Hybrid_Score"] = 0.6 * summary["SVD_Score"] + 0.4 * summary["Rule_Score"]

    # Sort top N
    summary = summary.sort_values(by="Hybrid_Score", ascending=False).head(top_n)

    return summary[["Airline", "Avg_Rating", "Rule_Score", "SVD_Score", "Hybrid_Score"]]

# ---------------------------
# Example Hybrid Recommendations
# ---------------------------
print("🎯 Hybrid Recommendations for Business Traveller")
print(hybrid_recommend("Business", top_n=5))

print("\n🎯 Hybrid Recommendations for Solo Leisure")
print(hybrid_recommend("Solo Leisure", top_n=5))
```
